```python
import tweepy
from tweepy import OAuthHandler

access_token = 'ACCESS_TOKEN'
access_token_secret = 'ACCESS_TOKEN_SECRET'
api_key = 'API_KEY
api_secret = 'API_SECRET'

auth = tweepy.OAuthHandler(api_key, api_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)

news_ids = ["@BBCWorld", "@CNN", "@nytimes", "ABC", "NBC News", "washingtonpost"]

class NewsListener(tweepy.StreamListener):
  def __init__(self, api):
      self.api = api

  def scrape_tweet (self, tweet):
    if tweet.user.id in news_ids:
      try:
        tweet.retweet

      except tweepy.TweepError as err:
        print(err.reason)  

def main():
  tweets_listener = NewsListener(api)
  stream = tweepy.Stream(api.auth, tweets_listener)
  stream.filter(track=["Japan"])

if __name__ == "__main__":
  main()
```
